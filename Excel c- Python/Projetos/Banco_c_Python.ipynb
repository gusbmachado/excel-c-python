```
###Análise do 80/20 -> Descobrir principal(ais) causas de cancelamento

import plotly.express as px #referência: https://plotly.com/python/

##criar figura
for coluna in tabela_clientes:
  fig = px.histogram(tabela_clientes, x=coluna, color="Categoria")
  #fig = px.histogram(tabela_clientes, x="Limite", color="Categoria") #Limite < ~10k ou ~15k -> Maior nº de cancelamentos

##exibir figura
  fig.show()
```

Da análise dos gráficos, percebe-se que:
1. Maioria dos clientes que cancelaram usavam conta blue
2. As chances de cancelamento aumentam conforme os clientes veem alguma nescecidade de entrar em contato (maior nº de ligações) -> Buscar atender àqueles que ligaram mais vezes

  * Mais de 2 a 3 chamadas

3. Principais razão dos cancelamentos são referentes ao limite do cartão e às transações. Ou seja, clientes que usam pouco o cartão tendem a cancelar o cartão, vide os dados:
